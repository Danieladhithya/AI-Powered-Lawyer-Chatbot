```python
# Install required libraries
!pip install PyPDF2 sentence-transformers numpy scikit-learn ipywidgets

# Import necessary libraries
from PyPDF2 import PdfReader
from sentence_transformers import SentenceTransformer
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import ipywidgets as widgets
from IPython.display import display, clear_output

# Load a pre-trained Sentence Transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')  # Lightweight and fast model

# Provide the paths of PDF files
pdf_paths = [
    r"/content/IPC Sections.pdf",
    r"/content/IPC part 2.pdf",
    r"/content/IPC part 3.pdf"
]

# Extract text from PDFs
raw_texts = []
for path in pdf_paths:
    pdf_reader = PdfReader(path)
    raw_text = ''
    for page in pdf_reader.pages:
        content = page.extract_text()
        if content:
            raw_text += content
    raw_texts.append(raw_text)

# Split text into smaller chunks
def split_text(text, chunk_size=100, overlap=20):  # Smaller chunks for better accuracy
    chunks = []
    start = 0
    while start < len(text):
        end = start + chunk_size
        chunks.append(text[start:end])
        start += (chunk_size - overlap)
    return chunks

texts = [split_text(raw_text) for raw_text in raw_texts]
flattened_texts = [item for sublist in texts for item in sublist]

# Generate embeddings for all text chunks
embeddings = model.encode(flattened_texts)

# Function to find the most relevant document
def find_most_relevant_doc(query, embeddings, texts):
    query_embedding = model.encode([query])
    similarities = cosine_similarity(query_embedding, embeddings)[0]
    most_relevant_index = np.argmax(similarities)
    return texts[most_relevant_index]

# Legal SOS Feature
def legal_sos():
    # Create a dropdown for user options
    options = widgets.Dropdown(
        options=["Contact a lawyer", "File a police complaint", "Connect with a helpline"],
        description="Select an option:",
        disabled=False,
    )

    # Create a button to submit the choice
    submit_button = widgets.Button(description="Submit", button_style="success")

    # Output area to display the result
    output_area = widgets.Output()

    # Function to handle button click
    def on_submit_button_click(b):
        with output_area:
            clear_output(wait=True)
            if options.value == "Contact a lawyer":
                print("Here are some nearby lawyers:")
                print("- Lawyer 1: +91-9876543210")
                print("- Lawyer 2: +91-9876543211")
                print("- Lawyer 3: +91-9876543212")
            elif options.value == "File a police complaint":
                print("To file a police complaint, visit the nearest police station or call 100.")
            elif options.value == "Connect with a helpline":
                print("Here are some helpline numbers:")
                print("- Women's Helpline: 1091")
                print("- Child Helpline: 1098")
                print("- Mental Health Helpline: 1800-599-0019")

    # Attach handler to button
    submit_button.on_click(on_submit_button_click)

    # Display the widgets
    display(widgets.VBox([options, submit_button, output_area]))

# DIY Legal Kits Feature
def diy_legal_kits():
    return """DIY Legal Kits:
1. Rental Agreement: Use our chatbot to generate a rental agreement.
2. Will: Create a will in minutes with our guided process.
3. Consumer Complaint: File a complaint using our step-by-step guide."""

# Main Bot Function
def get_bot_response(user_input):
    print(f"Debug: User input = {user_input}")  # Debugging statement
    bot_response = ""  # Initialize bot_response with an empty string

    # Clean the input
    user_input = user_input.strip().lower()

    if user_input in ["hi", "hello", "hey", "hy"]:
        bot_response = "Hello, welcome to Legal Chatbot. How can I assist you today!"
    elif user_input in ["bye", "by", "thank you", "thanks"]:
        bot_response = "Bye, and have a good day!"
    elif user_input == "legal sos":
        legal_sos()  # Call the enhanced Legal SOS feature
        return ""  # Return empty string to avoid printing anything else
    elif user_input == "diy legal kits":
        bot_response = diy_legal_kits()
    else:
        question = user_input  # Extract and clean the question from user input
        if len(question) < 4:
            bot_response = "Please enter a valid question!"
        else:
            try:
                # Find the most relevant document
                relevant_doc = find_most_relevant_doc(question, embeddings, flattened_texts)
                bot_response = f"Here is the relevant information:\n{relevant_doc}"
            except Exception as e:
                bot_response = f"An error occurred while processing your request: {e}"

    print(f"Debug: Bot response = {bot_response}")  # Debugging statement
    return bot_response

# Create UI Components
input_box = widgets.Text(placeholder="Type your question here...", layout=widgets.Layout(width="80%"))
send_button = widgets.Button(description="Send", button_style="success")
output_area = widgets.Output()

# Define Button Click Handler
def on_send_button_click(b):
    with output_area:
        clear_output(wait=True)  # Clear the output area
        user_input = input_box.value
        print(f"Debug: User input from UI = {user_input}")  # Debugging statement
        bot_response = get_bot_response(user_input)
        if bot_response:  # Only print if there's a response
            print(f"You: {user_input}")
            print(f"Bot: {bot_response}")
        input_box.value = ""  # Clear the input box after sending

# Attach Handler to Button
send_button.on_click(on_send_button_click)

# Display UI
display(widgets.VBox([
    widgets.HTML("<h1>Legal Chatbot</h1>"),
    widgets.HTML("<p>Type your question or use voice input:</p>"),
    input_box,
    send_button,
    output_area
]))

# Main Chatbot Loop (for non-widget environments)
print("Welcome to the Legal Chatbot!")
print("Type 'Legal SOS' for emergency help or 'DIY Legal Kits' for step-by-step guides.")
print("Type 'exit' to end the conversation.")

while True:
    user_input = input("You: ")
    if user_input.strip().lower() == "exit":
        print("Bot: Bye, and have a good day!")
        break
    bot_response = get_bot_response(user_input)
    if bot_response:  # Only print if there's a response
        print("Bot:", bot_response)
```
